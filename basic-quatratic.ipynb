```python
# Define the basic quadratic function
def g(x):
  return x**2 + 2*x + 5

# Plot the basic quadratic function
plt.plot(x, g(x), label='g(x)')

# Set the title and labels for the plot
plt.title("Basic Quadratic Function")
plt.xlabel("x")
plt.ylabel("y")

# Turn on the axis grids
plt.grid(True)

# Add a legend to the plot
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Generate a set of x-values using np.linspace ranging from -10 to 10
x = np.linspace(-10, 10, 100)
# Define the basic quadratic function
def g(x):
  return x**2 + 2*x + 5

# Plot the basic quadratic function
plt.plot(x, g(x), label='g(x)')

# Set the title and labels for the plot
plt.title("Basic Quadratic Function")
plt.xlabel("x")
plt.ylabel("y")

# Turn on the axis grids
plt.grid(True)

# Add a legend to the plot
plt.legend()

# Display the plot
plt.show()
```

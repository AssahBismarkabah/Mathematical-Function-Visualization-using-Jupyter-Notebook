```python
import numpy as np
import matplotlib.pyplot as plt
# Generate a set of x-values using np.linspace ranging from -10 to 10
x = np.linspace(-10, 10, 100)
# Define the function that combines the quadratic and sine functions
def g(x):
  return x**2 + 2*x + 5
    
def i(x):
  return np.sin(g(x))

# Plot the function that combines the quadratic and sine functions
plt.plot(x, i(x), label='i(x)')

# Set the title and labels for the plot
plt.title("Combined Quadratic and Sine Function")
plt.xlabel("x")
plt.ylabel("y")

# Turn on the axis grids
plt.grid(True)

# Add a legend to the plot
plt.legend()

# Display the plot
plt.show()

```

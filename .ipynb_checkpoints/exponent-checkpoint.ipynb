```python
import numpy as np
import matplotlib.pyplot as plt
# Generate a set of x-values using np.linspace ranging from -10 to 10
x = np.linspace(-10, 10, 100)
# Define the function that represents exponential decay
def j(x):
  return np.exp(-0.1*x)

# Plot the function that represents exponential decay
plt.plot(x, j(x), label='j(x)')

# Set the title and labels for the plot
plt.title("Exponential Decay Function")
plt.xlabel("x")
plt.ylabel("y")

# Turn on the axis grids
plt.grid(True)

# Add a legend to the plot
plt.legend()

# Display the plot
plt.show()
```
